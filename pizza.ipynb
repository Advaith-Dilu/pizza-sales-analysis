```python
# =============================
# üçï Pizza Sales EDA in Google Colab
# =============================

# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime

# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('Set2')

# Step 2: Upload File (you can use the Colab upload dialog)
from google.colab import files
uploaded = files.upload()

# Replace 'pizza_sales.csv' with your actual file name
df = pd.read_csv('pizza_sales.csv')

# Step 3: Basic Info
print("üîç Basic Dataset Overview")
print(df.shape)
print(df.info())
print(df.head())

# Step 4: Summary Statistics
print("\nüìä Statistical Summary:")
print(df.describe(include='all'))

# Step 5: Check Missing & Duplicates
print("\nüß© Missing Values:")
print(df.isnull().sum())

print("\nüìã Duplicate Rows:", df.duplicated().sum())

# Step 6: Data Cleaning (example)
# Convert date columns to datetime if present
date_cols = ['order_date', 'date', 'Order Date']
for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col])

# Step 7: Create useful columns if missing
if 'total_price' not in df.columns and {'quantity', 'price'}.issubset(df.columns):
    df['total_price'] = df['quantity'] * df['price']

# Step 8: Exploratory Data Analysis (Basic Insights)
print("\nüçï Top 5 Pizza Types by Sales:")
if 'pizza_name' in df.columns:
    print(df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(5))

print("\nüì¶ Orders per Pizza Category:")
if 'pizza_category' in df.columns:
    print(df['pizza_category'].value_counts())

# Step 9: Visualization ‚Äî Sales Overview
plt.figure(figsize=(8,5))
sns.histplot(df['total_price'], bins=30, kde=True)
plt.title('Distribution of Total Price')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.show()

# Step 10: Category-Wise Sales
if 'pizza_category' in df.columns:
    plt.figure(figsize=(8,5))
    category_sales = df.groupby('pizza_category')['total_price'].sum().sort_values(ascending=False)
    sns.barplot(x=category_sales.index, y=category_sales.values)
    plt.title('Total Sales by Pizza Category')
    plt.xlabel('Category')
    plt.ylabel('Total Sales ($)')
    plt.show()

# Step 11: Top 10 Best-Selling Pizzas
if 'pizza_name' in df.columns:
    top_pizzas = df.groupby('pizza_name')['total_price'].sum().sort_values(ascending=False).head(10)
    plt.figure(figsize=(10,6))
    sns.barplot(x=top_pizzas.values, y=top_pizzas.index)
    plt.title('Top 10 Best-Selling Pizzas')
    plt.xlabel('Total Sales ($)')
    plt.ylabel('Pizza Name')
    plt.show()

# Step 12: Sales Trend Over Time
if any(col in df.columns for col in date_cols):
    time_col = [col for col in date_cols if col in df.columns][0]
    daily_sales = df.groupby(df[time_col].dt.date)['total_price'].sum()
    plt.figure(figsize=(10,5))
    daily_sales.plot()
    plt.title('Daily Sales Trend')
    plt.xlabel('Date')
    plt.ylabel('Total Sales ($)')
    plt.show()

# Step 13: Correlation Analysis
numeric_cols = df.select_dtypes(include=[np.number])
plt.figure(figsize=(8,5))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='Blues')
plt.title('Correlation Heatmap')
plt.show()

# Step 14: Hourly/Day Analysis (if time available)
if 'order_time' in df.columns:
    df['order_hour'] = pd.to_datetime(df['order_time']).dt.hour
    plt.figure(figsize=(8,4))
    sns.countplot(x='order_hour', data=df)
    plt.title('Orders by Hour of the Day')
    plt.show()

print("\n‚úÖ EDA Completed Successfully!")

```
